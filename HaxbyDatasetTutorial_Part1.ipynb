# Haxby Dataset (from Nilearn database)
## Basic dataset loading and information printing

```python
# Numpy and Matplotlib library dependencies
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 

# Nilearn basic
from nilearn import datasets, plotting, image
```

```python
#Fetching the Haxby 2001 datasets(for 1 subject) of Nilearn database 
haxby_dataset = datasets.fetch_haxby()
```

```python
#now, The dataset (haxby_dataset) is accessible in many ways,
#use <Tab> for function choices and completion
#for example, to show dataset information
print(haxby_dataset)
```

```python
#Accessing the dataset
#Initializing some of the images
#Indexing in python start with 0, (while indexing in Matlab start with 1)
#Example, If the dataset consist of 7 subjects, the last subject index is 6
haxby_anat_sub1 = haxby_dataset.anat[0] # Subject 1 anatomical image
haxby_mask_sub1 = haxby_dataset.mask_vt [0]
haxby_BOLD_sub1 = haxby_dataset.func[0]
```

```python
#Print basic info about the dataset
print('Subject 1 anatomical nifti image(3D) is stored at: %s' %haxby_dataset.anat[0])
print('Subject 1 functional(BOLD) nifti image(4D) is stored at: %s' %haxby_dataset.func[0])
```

# Plotting and showing the dataset

Using anatomical maps of Subject 1 dataset

```python
# Plotting the anatomical maps of Subject 1 (), ortho display mode
plotting.plot_anat(haxby_anat_sub1, title="Anatomical Maps Plotting")
```

```python
#for different cut coordinates 
coordinates = (10,10,10)
plotting.plot_anat(haxby_anat_sub1, cut_coords = coordinates, title="Anatomical Maps Plotting")
```

Observe the difference of images between two different coordinates.

This plot has plotted for two different slices of Subject 1 fMRI datasets

```python
# For Saggital(x) display mode
plotting.plot_anat(haxby_anat_sub1, display_mode = 'x', title="Saggital Anatomical Maps Plotting")
```

```python
# For Coronal(y) display mode
plotting.plot_anat(haxby_anat_sub1, display_mode = 'y', title="Coronal Anatomical Maps Plotting")
```

```python
# For Axial(z) display mode
plotting.plot_anat(haxby_anat_sub1, display_mode = 'z', title="Axial Anatomical Maps Plotting")
```

```python
# Saving image of slices into folder in .png filetype
# the image will be saved in the ipython folder without showing in this notebook
plotting.plot_anat(haxby_anat_sub1, output_file= "haxby_anat_sub1_saggital.png", display_mode= 'x', title= "Saggital")
```
